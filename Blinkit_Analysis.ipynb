# **Data Analysis - Blinkit Analysis **

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('/content/blinkit_data.csv')
df.head()
```

```python
df.head()
```

```python
df['Item Fat Content'] = df['Item Fat Content'].str.replace('LF','Low Fat').str.replace('low fat','Low Fat').str.replace('reg','Regular')
```

```python
df.head(10)
```

```python
df.columns
```

### **Business Requirments**

KPI's Requiremnts

```python
# Total sales
Total_sales = df['Sales'].sum()
print(Total_sales)
```

```python
# Average Sales
Average_sales = df['Sales'].mean()
print(Average_sales)
```

```python
# No. of Items Sold
No_of_items_sold = df['Sales'].count()
print(No_of_items_sold)
```

```python
# Average Rating
avg_rating = df['Rating'].mean()
print(avg_rating)
```

```python
print(f"Total Sales: {Total_sales:.1f}")
print(f"Average Sales: {Average_sales:.1f}")
print(f"No. of Items Sold: {No_of_items_sold:.0f}")
print(f"Average Rating: {avg_rating:.1f}")
```

**Charts Requirements**

```python
# Total Sales by Fat Content
total_sales_by_fat_content = df.groupby('Item Fat Content')['Sales'].sum()
plt.pie(total_sales_by_fat_content,labels=total_sales_by_fat_content.index
        ,autopct='%.1f%%',startangle=90)
plt.title('Total Sales by Fat Content')
plt.show()
```

```python
# Total sales by item type
plt.figure(figsize=(10,6))
total_sales_by_item_type = df.groupby('Item Type')['Sales'].sum()
plt.bar(total_sales_by_item_type.index,total_sales_by_item_type.values)
plt.xticks(rotation=90)
plt.xlabel('Item Type')
plt.ylabel('Total Sales')
plt.title('Total Sales by Item Type')
plt.tight_layout()
plt.show()
```

```python
grouped = df.groupby(['Outlet Location Type','Item Fat Content'])['Sales'].sum().unstack()
grouped = grouped[['Regular','Low Fat']]
ax = grouped.plot(kind='bar',figsize=(10,6))
plt.xlabel('Outlet Location Type')
plt.ylabel('Total Sales')
plt.title('Total Sales by Outlet Location Type and Item Fat Content')
```

```python
# Total sales by outlet Establishment
plt.figure(figsize=(10,6))
total_sales_by_outlet_establishment = df.groupby('Outlet Establishment Year')['Sales'].sum()
plt.plot(total_sales_by_outlet_establishment.index,total_sales_by_outlet_establishment.values)
```

```python
# Sales by outlet_size

sales_by_size = df.groupby('Outlet Size')['Sales'].sum()
plt.figure(figsize=(10,6))
plt.pie(sales_by_size,labels=sales_by_size.index,autopct='%.1f%%',startangle=90)
plt.title('Outlet Size')
plt.tight_layout()
plt.show()
```

```python
# Sales by outler location
sales_by_location = df.groupby('Outlet Location Type')['Sales'].sum()
plt.figure(figsize=(8,6))
ax = sns.barplot(y=sales_by_location.index,x=sales_by_location.values)
plt.xlabel('Outlet Location Type')
plt.ylabel('Total Sales')
plt.title('Total Sales by Outlet Location Type')
plt.tight_layout()
plt.show()
```
